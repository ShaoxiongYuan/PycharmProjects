```python
import sklearn.datasets as sd
import sklearn.utils as su
import sklearn.metrics as sm
import sklearn.model_selection as ms
import sklearn.ensemble as se
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
sns.heatmap(data.isna(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isna(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age


data['Age'] = data[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
data.drop('Cabin', axis=1, inplace=True)
data.dropna(inplace=True)
sex = pd.get_dummies(data['Sex'], drop_first=True)
embark = pd.get_dummies(data['Embarked'], drop_first=True)
train = pd.concat([data, sex, embark], axis=1)
train.drop(['Sex', 'Embarked'], axis=1, inplace=True)
train = train.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
train.head()
```

```python
sns.heatmap(data.isna(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
X = train.drop('Survived', axis=1)
y = data['Survived']

model = se.RandomForestClassifier(n_estimators=500,
                                  max_depth=7,
                                  min_samples_split=3)
model.fit(X, y)
```

```python
data = pd.read_csv('test.csv')
data.head()
```

```python
data['Age'] = data[['Age', 'Pclass']].apply(impute_age, axis=1)
data.drop('Cabin', axis=1, inplace=True)
data.dropna(inplace=True)
sex = pd.get_dummies(data['Sex'], drop_first=True)
embark = pd.get_dummies(data['Embarked'], drop_first=True)
test = pd.concat([data, sex, embark], axis=1)
test.drop(['Sex', 'Embarked'], axis=1, inplace=True)
test = test.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
test.head()
```

```python
predictions = model.predict(test)
len(predictions)
```

```python
data = pd.read_csv('test.csv')
data.head(1)
```

```python
data['Survived'] = pd.Series(predictions, index=test.index)
data
```

```python
data.to_csv('result.csv')
```
