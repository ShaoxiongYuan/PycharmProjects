```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Salary_Data.csv')
data.head()
```

```python
data.plot(x='YearsExperience', y='Salary', kind='scatter')
```

```python
x, y = data['YearsExperience'], data['Salary']
w0, w0s = 1, []
w1, w1s = 1, []
lrate = 0.001  # 
epoch = 1000  # for
epoches = []
losses = []

for i in range(epoch):
    # 
    epoches.append(i + 1)
    w0s.append(w0)
    w1s.append(w1)
    loss = ((w0 + w1 * x - y)**2).sum() / 2
    losses.append(loss)
    print('{:4}> w0={:.8f}, w1={:.8f}, loss={:.8f}'.format(
        i + 1, w0, w1, loss))

    # w0,w1
    d0 = (w0 + w1 * x - y).sum()
    d1 = (x * (w0 + w1 * x - y)).sum()
    w0 = w0 - lrate * d0
    w1 = w1 - lrate * d1

predictions = w0 + w1 * x
plt.grid(ls=':')
plt.scatter(x, y)
plt.plot(x, predictions, color='orangered')
```
