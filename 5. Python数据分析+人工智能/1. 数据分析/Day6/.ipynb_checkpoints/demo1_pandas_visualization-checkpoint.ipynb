```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
ts = pd.Series([1, 2, 5, 2, 6])
ts.cumsum()
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = np.arange(len(df3))
df3.plot(x='A', y='B')
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3.plot(kind='scatter', x='B', y='C')
```

```python
s = np.random.normal(75, 3, 10)
s = pd.Series(s)
s.plot(kind='hist', bins=5)
```

```python
s.plot(kind='pie')
import matplotlib.pyplot as plt
plt.axis('equal')
plt.legend()
```

```python
ts = pd.Series([1, 4, 8, 3, 8], index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
ts.plot(kind='bar')
```

```python
ts1 = pd.Series([1, 4, 8, 3, 8], index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
ts2 = pd.Series([4, 2, 6, 2, 1], index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
df = pd.DataFrame({'ts1': ts1, 'ts2': ts2})
df.plot.bar()
```

```python
plt.axis('equal')
ts.plot.pie(autopct='%d%%')
plt.legend()
```

```python
df.plot.pie(subplots=True, figsize=(12, 6), autopct='%d%%')
```

```python
df.plot.scatter(x='ts1',
                y='ts2',
                s=[50, 80, 100, 120, 150, 180],
                cmap='jet',
                c='ts2')
```

```python
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.plot.box(vert=False)
```

```python
df = pd.DataFrame(np.random.rand(10, 2), columns=['Col1', 'Col2'])
df['X'] = pd.Series(['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'])
df.boxplot(by='X')
```
