```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('aapl.csv',
                   header=None,
                   usecols=[1, 3, 4, 5, 6, 7],
                   names=['date', 'open', 'high', 'low', 'close', 'volume'])
```

```python
def dmy2ymd(dmy):
    return '-'.join(dmy.split('-')[::-1])


data['date'] = data['date'].apply(dmy2ymd)
data['date'] = data['date'].astype('M8[D]')
data.info()
```

```python
data.plot(x='date', y='close', figsize=(10, 5))
```

```python
# 整理日期,变为数字字段
days = (data['date'] - pd.to_datetime('2011-1-27')).dt.days
A = np.column_stack((days, np.ones_like(days)))
B = data['close']
x = np.linalg.lstsq(A, B)
x[0]
```

```python
trend_line = x[0][0] * days + x[0][1]
data['trend'] = trend_line
data.plot(x='date', y=['trend', 'close'])
```
