````python
desc = sub_data.describe()
desc
````

````python
corr = sub_data.corr()
````

````python
corr[corr > 0.4] = 9999
corr[corr < 0.4] = 0
````

````python
corr
````

````python
mean_score = desc.loc['mean'].sort_values(ascending=False)
fields = pd.read_csv(filepath_or_buffer='health_headers_intro.csv',
                     encoding='gbk',
                     header=None,
                     names=['fid', 'effect', '_'],
                     usecols=['fid', 'effect'],
                     index_col='fid')
fields
````

````python
fields['mean_score'] = mean_score
````

````python
fields
````

通过上表可得,多数产品在产品功效上更多关注用户的精神状态,抗皮肤衰老,再次为身体健康,最后为外界影响

## 每种保健品的频数

````python
products = {
    1: '太太**',
    2: '乌鸡**',
    3: '排毒**',
    4: '阿*',
    5: '红桃*',
    6: '脑白*',
    7: '朵**',
    9: '其他'
}
products = pd.Series(products)
result = pd.DataFrame({'products': products, 'q8': data['q8'].value_counts()})
result
````

有结果可知,市场占有率最高的是'排毒'

## 研究每个产品功效的侧重点

````python
result1 = pd.DataFrame()
for k, v in products.items():
    mask = data['q8'] == k
    result1[k] = sub_data[mask].mean()
result1
````

````python
# 获取每一列最大值索引
result_idxmax = result1.idxmax()
result_idxmax
````

````python
result1 = pd.DataFrame({'product': products, 'qid': result_idxmax})
result1
````

````python
fields.sort_values('mean_score', ascending=False)
````

````python
def func(x):
    return fields.loc[x]['effect']


result1['effect'] = result1['qid'].apply(func)
result1
````

#### 每个产品功效如图所示:
若希望切入保健品市场,优先保证精神状态

## 研究每个产品已婚未婚频数统计

````python
result3 = pd.DataFrame()
for k, v in products.items():
    mask = data['q8'] == k
    result3[k] = data[mask]['w1'].value_counts()
result3 = result3.T
result3.columns = ['Married', 'Not-Married']
result3['product'] = products
result3['effect'] = result1['effect']
result3
````
