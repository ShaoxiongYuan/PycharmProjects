```python
plt.hlines(
    [2, 3, 3.5],  # 垂线的x坐标值
    [10, 20, 30],  # 每条垂线起始y坐标
    [25, 35, 45])  # 每条垂线结束y坐标
```

```python
plt.vlines(
    [2, 3, 3.5],  # 垂线的x坐标值
    [10, 20, 30],  # 每条垂线起始y坐标
    [25, 35, 45])  # 每条垂线结束y坐标
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# 绘制图形
plt.plot(x, y1, label="sin", linewidth=2)  # 实线，线宽2像素
plt.plot(x, y2, label="cos", linestyle="--", linewidth=4)  # 虚线，线宽4像素
plt.plot(x, y3, label='tan')
plt.legend(loc='upper right')
plt.title('sin & cos')
plt.xlim(-2, 2)
plt.ylim(-3, 3)
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x) / 2

plt.plot(x, y1, c='green', label=r'$y=sin(x)$', ls='-', lw=4, alpha=0.7)
plt.plot(x, y2, c='magenta', label=r'$y=\frac{1}{2}cos(x)$', ls='-.', lw=2)
plt.legend(loc='lower right')
# 设置刻度
vals = [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]
text = [r'$-\pi$', r'$-\frac{\pi}{2}$', '0', r'$\frac{\pi}{2}$', r'$\pi$']
plt.xticks(vals, text)

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

plt.yticks([-1, -0.5, 0.5, 1])

xs = [np.pi / 2, np.pi / 2]
ys = [0, 1]
plt.scatter(xs,
            ys,
            marker='o',
            s=50,
            facecolor='green',
            edgecolor='red',
            zorder=3)

# 添加备注
plt.annotate(r'$(\frac{\pi}{2},1)$',
             xycoords='data',
             xy=(np.pi / 2, 1),
             textcoords='offset points',
             xytext=(30, 20),
             fontsize=14,
             arrowprops=dict(arrowstyle='->', connectionstyle='angle3'))
```
