```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
ts = pd.Series([1, 2, 5, 2, 6])
ts.cumsum()
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = np.arange(len(df3))
df3.plot(x='A', y='B')
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3.plot(kind='scatter', x='B', y='C')
```

```python
s = np.random.normal(75, 3, 10)
s = pd.Series(s)
s.plot(kind='hist', bins=5)
```

```python
s.plot(kind='pie')
import matplotlib.pyplot as plt
plt.axis('equal')
plt.legend()
```
