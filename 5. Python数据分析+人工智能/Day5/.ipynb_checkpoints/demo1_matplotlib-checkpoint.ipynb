```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.array([1, 2, 3, 8, 10])
y = np.array([5, 1, 7, 12, 19])

plt.plot(x, y)
```

```python
plt.hlines(
    [2, 3, 3.5],  # 垂线的x坐标值
    [10, 20, 30],  # 每条垂线起始y坐标
    [25, 35, 45])  # 每条垂线结束y坐标
```

```python
plt.vlines(
    [2, 3, 3.5],  # 垂线的x坐标值
    [10, 20, 30],  # 每条垂线起始y坐标
    [25, 35, 45])  # 每条垂线结束y坐标
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# 绘制图形
plt.plot(x, y1, label="sin", linewidth=2)  # 实线，线宽2像素
plt.plot(x, y2, label="cos", linestyle="--", linewidth=4)  # 虚线，线宽4像素
plt.plot(x, y3, label='tan')
plt.legend(loc='upper right')
plt.title('sin & cos')
plt.xlim(-2, 2)
plt.ylim(-3, 3)
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x) / 2

plt.plot(x, y1, c='green', label='sin', ls='-', lw=4, alpha=0.3)
plt.plot(x, y2, c='magenta', label=r'cos/2', ls='-.', lw=2)
plt.legend(loc='lower right')
# 设置刻度
vals = [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]
text=[r'$-\pi$',r'$-\frac{\pi}{2}$','0',r'$\frac{\pi}{2}$',r'$\pi$']
plt.xticks(vals,text)
```
