```python
import numpy as np
import pandas as pd
import gspread
from google.oauth2.service_account import Credentials
from gspread_pandas import Spread, Client
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
stocks_df = pd.read_html(url, header=0)[0]
stocks_df.head()
```

```python
stocks_df.info()
```

```python
plt.figure(figsize=(12, 7))
stocks_df['GICS Sector'].value_counts().sort_values(ascending=False).plot(
    kind='bar', rot=30)
```

```python
plt.figure(figsize=(12, 7))
stocks_df['GICS Sub Industry'].value_counts().sort_values(
    ascending=False).head(10).plot(
        kind='bar', rot=30)
```

```python
stocks_df['Price_1_1'] = stocks_df['Symbol'].apply(
    lambda x: '=INDEX(GOOGLEFINANCE("' + x + '","Price", "1/1/2020"),2,2)')
stocks_df['Price_3_23'] = stocks_df['Symbol'].apply(
    lambda x: '=INDEX(GOOGLEFINANCE("' + x + '","Price", "3/23/2020"),2,2)')
stocks_df['Price_4_9'] = stocks_df['Symbol'].apply(
    lambda x: '=INDEX(GOOGLEFINANCE("' + x + '","Price", "4/9/2020"),2,2)')
```

```python
stocks_df['Shares'] = stocks_df['Symbol'].apply(lambda x: '=GOOGLEFINANCE("' +
                                                x + '","shares")')
```

```python
scope = [
    'https://spreadsheets.google.com/feeds',
    'https://www.googleapis.com/auth/drive'
]
credentials = Credentials.from_service_account_file(
    'credentials/credentials.json', scopes=scope)
```

```python
client = Client(scope=scope, creds=credentials)
spread = Spread('stocks-data', client=client)
```

```python
cols_to_keep = [
    'Symbol', 'Security', 'GICS Sector', 'GICS Sub Industry', 'Price_1_1',
    'Price_3_23', 'Price_4_9', 'Shares'
]
```

```python
spread.df_to_sheet(stocks_df[cols_to_keep])
```

```python
stocks_df = spread.sheet_to_df()
stocks_df.head()
```

```python
for index, values in stocks_df.iteritems():
    for i,v in values.items():
        if '#N/A' in v:
            stocks_df.drop(i, inplace=True)
```

```python
stocks_df[['Price_1_1', 'Price_3_23', 'Price_4_9', 'Shares']] = stocks_df[[
    'Price_1_1', 'Price_3_23', 'Price_4_9', 'Shares'
]].apply(pd.to_numeric)
```

```python
stocks_df['Marketcap_1_1'] = stocks_df['Price_1_1'] * stocks_df['Shares']
stocks_df['Marketcap_3_23'] = stocks_df['Price_3_23'] * stocks_df['Shares']
stocks_df['Marketcap_4_9'] = stocks_df['Price_4_9'] * stocks_df['Shares']
```

```python
stocks_df['PercentageChange_3_23_1_1'] = (
    stocks_df['Price_3_23'] -
    stocks_df['Price_1_1']) / stocks_df['Price_1_1'] * 100
stocks_df['PercentageChange_4_9_3_23'] = (
    stocks_df['Price_4_9'] -
    stocks_df['Price_3_23']) / stocks_df['Price_3_23'] * 100
stocks_df['PercentageChange_4_9_1_1'] = (
    stocks_df['Price_4_9'] -
    stocks_df['Price_1_1']) / stocks_df['Price_1_1'] * 100
```

```python
sum(stocks_df['Marketcap_3_23'] - stocks_df['Marketcap_1_1']) / 10**9
```

```python
sum(stocks_df['Marketcap_4_9'] - stocks_df['Marketcap_3_23']) / 10**9
```

```python
sum(stocks_df['Marketcap_4_9'] - stocks_df['Marketcap_1_1']) / 10**9
```

```python
plt.figure(figsize=(18, 6))
plt.tick_params('both', labelsize='8')
plt.xticks(rotation=45)
sns.boxplot(
    x=stocks_df['GICS Sector'], y=stocks_df['PercentageChange_4_9_1_1'])
```
